```python
income_dict = {}

for row in range(len(efficiency_df.index)):
    country = efficiency_df.iloc[int(row)]['Country']
    income = efficiency_df.iloc[int(row)]['Income group classification']
    income_dict[country] = income
    
income_dict
```

```python
system_df['Income group classification'] = [income_dict[x] if x in income_dict.keys() else 'No income group' for x in system_df['Country']] 
equity_df['Income group classification'] = [income_dict[x] if x in income_dict.keys() else 'No income group' for x in equity_df['Country']]
```
